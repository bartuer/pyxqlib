```python
from pyxqlib import Tsidx
from pyxqlib import MustelasQuote
```

```python
q = pd.read_pickle(f"data/quote_df.pkl")
ii = Tsidx()
jj = 0                                                                                                                                                          
for s, v in q.groupby(level="instrument"):                                                                                                                 
    d = v.droplevel(level="instrument")                              
    ii[jj] = d.index.values.astype('datetime64[s]').astype('uint32')                                                                                                      
    jj += 1 
idx = ii[0]
```

```python
length, start, end, days = (len(idx), idx.start, idx.stop, idx.dlen)
```
